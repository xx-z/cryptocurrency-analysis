```python
#yfinanceとpandasをインポート
import yfinance as yf
import pandas as pd

#取得するデータ詳細
name = 'BTC-USD'
start = '2022-01-01'
end = '2023-01-01'

#データのダウンロード
df = yf.download(tickers=name, start=start, end=end)
df.head()
```

```python
#1日ごとの終値データを表示
df["Close"].plot()
```

```python
#時間を指定してみる
import datetime as dt
df[df.index >= dt.datetime(2022, 6, 1)]["Close"].plot()
```
